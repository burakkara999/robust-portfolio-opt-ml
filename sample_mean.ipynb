```python
import pandas as pd
import numpy as np
from asset_data_module import read_close_prices_all_merged, save_log_returns
from features import make_feature_windows
```

```python
# markets = ['dow30', 'commodities', 'bonds']
markets = ['crypto', 'commodities', 'bonds']
markets_text = "crypto-commodities-bonds"
# markets = ['dow30']
# markets = ['commodities']
# markets = ['bonds']
start_date, end_date = "2022-01-01", "2025-11-28"
# start_date, end_date = "2024-06-01", "2025-11-28"

_, close_df = read_close_prices_all_merged(markets, after_date=start_date)
close_df = close_df.loc[:end_date]

##SAVE LOG RETURNS
save_log_returns(close_df, out_path='data/log_returns/log_returns_crypto-commodities-bonds.csv', lag=2)

rolling = make_feature_windows(
    close_prices=close_df,
    lookback=60,
    horizon=1,
    days_per_week=2
)
# rolling[0]['past_prices']
```

```python
returns_df = pd.read_csv(f"data/log_returns/log_returns_{markets_text}.csv")

returns_df = returns_df.sort_values("Date").set_index("Date")
```

```python
##log returns SM60
rolling_mean_60 = returns_df.shift(1).rolling(window=60, min_periods=60).mean()

##Residuals
residuals_sm60 = returns_df - rolling_mean_60

# drop the initial rows where SM60 is NaN (first ~60 rows)
residuals_sm60 = residuals_sm60.dropna(how="all")
residuals_sm60.to_csv(f"data/prediction/SM60_{markets_text}_residuals.csv")

# rolling_mean_60.head(61)
rolling_mean_60_clean = rolling_mean_60.dropna(how="all")
rolling_mean_60_clean.to_csv(f"data/prediction/SM60_{markets_text}_log_returns.csv")
# rolling_mean_60_clean
```

```python
##log returns SM30
rolling_mean_30 = returns_df.shift(1).rolling(window=30, min_periods=30).mean()

##Residuals
residuals_sm30 = returns_df - rolling_mean_30

# drop the initial rows where SM30 is NaN (first ~30 rows)
residuals_sm30 = residuals_sm30.dropna(how="all")
residuals_sm30.to_csv(f"data/prediction/SM30_{markets_text}_residuals.csv")

# rolling_mean_30.head(61)
rolling_mean_30_clean = rolling_mean_30.dropna(how="all")
rolling_mean_30_clean.to_csv(f"data/prediction/SM30_{markets_text}_log_returns.csv")
# rolling_mean_30_clean
```
